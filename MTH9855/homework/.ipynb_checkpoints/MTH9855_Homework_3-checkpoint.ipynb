### Problem 1

Consider optimally trading a single stock over $T = 30$ days.
Each period is one day, and you can trade once per day. The stockâ€™s daily
return volatility is $\sigma$. Suppose your forecast is 50 basis points for the first
period, and decays exponentially with half-life 5 days. This means that
$$
\alpha_t := \mathbb{E}[r_{t,t+1}] = 50\times 10^{-4}\times 2^{-t/5}.
$$
Let $c(\delta)$ be the cost, in dollars, of trading $\delta$ dollars of this stock. For selling,
$\delta< 0$. Following Almgren, we assume that
$$
c(\delta) = PX\left[ \frac{\gamma\sigma}{2}\frac{X}{V}\left(\frac{\Theta}{V}\right)^{1/4} + \text{sign}(X)\eta\sigma\left| \frac{X}{V} \right|^\beta\right], \quad X =\delta/P
$$
where $P$ is the current price in dollars, $X$ is the signed trade size in shares, $V$ is the daily volume in shares, $\Theta$ is the total number of shares outstanding, and finally $\gamma=0.314$ and $\eta=0.142$ and $\beta=0.6$ are constants fit to market data. For concreteness, suppose the asset we are trading has
$$
P =\$ 40, V = 2\times 10^6, \Theta=2\times 10^8, \sigma=0.02.
$$
For a trading path $\mathbf{x} = (x_0 , x_1 , \cdots, x_T )$ where $x_t$ denotes dollar holdings of
the stock at time $t$, define the profit (also in dollars) as
$$
\pi(\mathbf{x}) = \sum_t \left[ x_t r_{t,t+1} - c(x_t - x_{t-1}) \right].
$$
This is a random variable due to the presence of $r_{t,t+1}$ which you can assume
is Gaussian with mean $\alpha_t := \mathbb{E}[r_{t,t+1} ]$ and variance $\mathbb{V}[r_{t,t+1} ] = \sigma^2$ . In this problem, always assume $x_0 = 0$ is fixed.

(a) Find the sequence of positions $x_1,x_2,\cdots,x_T$ that maximizes
$$
u(x_1,\cdots,x_T) = \sum_{t=1}^{T}\left[ x_t\alpha_t -\frac{\kappa}{2}\sigma^2 x_t^2 -c(x_t-x_{t-1}) \right]
$$
with risk-aversion $\kappa=10^{-7}$. Set tolerance so that your algorithm does not terminate unless each $x_t\in \mathbb{R}$ is within a distance of one dolloar to the true optimal path. Plot the optimal path $\mathbf{x}^* := (x_0=0, x_1^*,\cdots, x_T^*)$ and also report its values in a table. Also report the computation time. Submit your code and a clear explanation of the algorithm you used, why your chose it over other possible algorithms, and how you know that it converges. For example, if you used a method that requires convexity, explain why the function you are optimizing is convex.

#### Solution

```python
gamma = 0.314
eta = 0.142
beta = 0.6
P = 40.0
V = 2000000.0
Theta = 200000000.0
sigma = 0.02
kappa = 0.0000001 # risk aversion
tau = 5.0 # half-life
alpha = 50.0 # initial alpha strength
T = 30 # Total time steps
```

```python
c0 = 0.5*kappa*sigma*sigma
c1 = gamma*sigma*pow(Theta/V,0.25)/(2.0*V*P)
c2 = eta*sigma/pow(P*V,beta)
print("c0 = ", c0)
print("c1 = ", c1)
print("c2 = ", c2)
```

Thus, $\frac{1}{2}\kappa\sigma^2 \approx 1.4\times 10^{-10}$, $c(\delta) = c_1 \delta^2 + c_2 |\delta|^{1.6}$, where $c_1 \approx 1.2412\times 10^{-10}, c_2 \approx 5.1459\times 10^{-8}$. To maximize,
$$
u(x_1,\cdots,x_T) = \sum_{t=1}^{T}\left[ \alpha_t x_t -c_0x_t^2 -c_1(x_t-x_{t-1})^2 
- c_2 \left|x_t-x_{t-1}\right|^{1.6}\right],
$$
where $\alpha_t = 0.005\times 2^{-t/5}$. Write $\delta_t = x_t -x_{t-1}$, $x_t = x_0 + \sum_{s=1}^{t}\delta_s$, where $x_0=0$ by assumption, 
$$
f(\delta_1,\cdots,\delta_T) = -u(x_1,\cdots,x_T) =  -\sum_{t=1}^{T}\alpha_t \sum_{s=1}^{t}\delta_s + c_0 \sum_{t=1}^{T}\left(  \sum_{s=1}^{t}\delta_s \right)^2 + c_1 \sum_{t=1}^{T}\delta_t^2 + c_2 \sum_{t=1}^{T}|\delta_t|^{1.6}.
$$
The goal is to find the optimal single-asset trading path. Notice that as a multi-variable function of $(\delta_1,\cdots,\delta_T)$,

- $f(\delta_1,\cdots,\delta_T)$ consists of terms of linear in $\delta$'s, quadratic in $\delta$'s with positive coefficients, and $\delta$'s raised to power 1.6 with positive coefficients. Thus, all terms in $f(\delta_1,\cdots,\delta_T)$ are convex.

- The only non-differntiable terms in $f(\delta_1,\cdots,\delta_T)$: $\sum_{t=1}^{T}|\delta_t|^{1.6}$ is separable.

As a consequence, Tseng's theorem apply and the algorithm of coordinate descent over the trades $\delta_1,\cdots,\delta_T$ is guaranteed to converge. Choose initial condition $\mathbf{x} = (x_0 , x_1 , \cdots, x_T ) = (0,0,\cdots,0)$. In other words, $(\delta_1 , \cdots, \delta_T ) = (0,\cdots,0)$. Then repeately iterate cyclically through $t=1,\cdots,T$, and perform the following optimization and update:
$$
\delta_t = \text{argmin}_\omega f(\delta_1,\cdots,\delta_{t-1},\omega,\delta_{t+1},\cdots,\delta_T),
$$
while keeping all other trades fixed. For a single-parameter optimization with respect to $\delta_t$, the relevant part is
\begin{eqnarray}
f(\delta_1,\cdots,\delta_T) &=& - \left(\sum _ {s=t}^T \alpha_s\right) \delta_t + c_0(T-t+1)\delta_t^2 + 2 c_0\left(\sum _ {s=t}^T\sum _ {r=1(\neq t)}^s \delta_r\right)\delta_t + c_1 \delta_t^2 + c_2 |\delta_t|^{1.6} + \text{irrelevant terms}\\
&=& \left[c_0(T-t+1) + c_1\right]\delta_t^2 + c_2 |\delta_t|^{1.6} +  \sum _ {s=t}^T \left( 2c_0\sum _ {r=1(\neq t)}^s \delta_r - \alpha_s\right)  \delta_t + \text{irrelevant terms}.
\end{eqnarray}


```python
from scipy.optimize import minimize_scalar
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import time
```

```python
# single parameter function to optimize
def func(x, t, T, c0, c1, c2, beta, tau, alpha, deltas):    
    coeff_1 = c0*(T-t+1)+c1
    coeff_2 = c2
    
    coeff_3 = 0
    for s in range(t,T+1):
        coeff_4 = 0
        for r in range(1,s+1):
            if r != t:
                coeff_4 += deltas[r-1]
        coeff_4 *= (2*c0)
        coeff_4 -= 0.0001*alpha/pow(2,s*1.0/tau)
        coeff_3 += coeff_4
    
    return coeff_1*x*x + coeff_2*pow(abs(x),1+beta) + coeff_3*x        
```

```python
def calc_pnl_and_sharpe(T, sigma, c1, c2, beta, tau, alpha, deltas, positions):
    
    pnl_mean = 0.0
    for t in range(1,T+1):
        revenue = 0.0001*alpha/pow(2,t*1.0/tau)*positions[t-1]
        cost = c1*deltas[t-1]*deltas[t-1] + c2*pow(abs(deltas[t-1]),1+beta)
        pnl_mean += (revenue-cost)
        
    pnl_var = 0.0
    for t in range(1,T+1):
        pnl_var += positions[t-1]*positions[t-1]*sigma*sigma
    
    return (pnl_mean, np.sqrt(pnl_var))
```

```python
def run(gamma, eta, beta, P, V, Theta, sigma, kappa, tau, alpha, T):
    c0 = 0.5*kappa*sigma*sigma
    c1 = gamma*sigma*pow(Theta/V,0.25)/(2.0*V*P)
    c2 = eta*sigma/pow(P*V,beta)
    
    # Initial guess (all zeros)
    deltas = [0]*T
    positions = np.cumsum(deltas)
    
    # Iteration control
    it=0
    max_it=10
    diff = 10
    
    while it<max_it and diff>1:
        for t in range(1,T+1):
            args = (t,T,c0,c1,c2,beta,tau,alpha,deltas)
            res = minimize_scalar(func, method='brent', args=args)
            deltas[t-1] = res.x
        new_positions = np.cumsum(deltas)
        diff = np.max(np.absolute(new_positions - positions))
        positions = new_positions
        it += 1
        
    (pnl_mean, pnl_std) = calc_pnl_and_sharpe(T, sigma, c1, c2, beta, tau, alpha, deltas, positions)
        
    return (deltas, positions, pnl_mean, pnl_std)
```

```python
ts = time.time()
(deltas, positions, pnl_mean, pnl_std) = run(gamma, eta, beta, P, V, Theta, sigma, kappa, tau, alpha, T)
te = time.time()
print("Computation time:", "%.6f" % (te-ts), ' sec')
print("PnL Mean =", pnl_mean)
print("PnL Std =", pnl_std)
print("Sharpe =", np.sqrt(252)*pnl_mean/pnl_std)
```

```python
plt.rcParams["figure.figsize"] = [15,7]
f, axarr = plt.subplots(1, 2)
axarr[0].plot(deltas)
axarr[0].set_title('Trades')
axarr[0].set_xlabel('Days')
axarr[1].plot(positions)
axarr[1].set_title('Positions')
axarr[1].set_xlabel('Days')
axarr[0].grid(True)
axarr[1].grid(True)
plt.show()
```

**(1) PnL and Sharpe ratio as a function of $\kappa$**:

```python
gamma = 0.314
eta = 0.142
beta = 0.6
P = 40
V = 2000000
Theta = 200000000
sigma = 0.02
kappa = 0.0000007 # risk aversion
tau = 5.0 # half-life
alpha = 50.0 # initial alpha strength
T = 30 # Total time steps

kappa_list = np.linspace(0.0000001, 0.000001, 20)
pnl_list = []
sharpe_list = []
for kappa in kappa_list:
    (deltas, positions, pnl_mean, pnl_std) = run(gamma, eta, beta, P, V, Theta, sigma, kappa, tau, alpha, T)
    pnl_list.append(pnl_mean)
    sharpe_list.append(np.sqrt(252)*pnl_mean/pnl_std)
    
plt.rcParams["figure.figsize"] = [15,7]
f, axarr = plt.subplots(1, 2)
axarr[0].plot(kappa_list, pnl_list)
axarr[0].set_title('PnL')
axarr[0].set_xlabel('Kappa')
axarr[1].plot(kappa_list, sharpe_list)
axarr[1].set_title('Sharpe')
axarr[1].set_xlabel('Kappa')
axarr[0].grid(True)
axarr[1].grid(True)
plt.show()
```

**(2) PnL and Sharpe ratio as a function of half-life $\tau$**:

```python
gamma = 0.314
eta = 0.142
beta = 0.6
P = 40
V = 2000000
Theta = 200000000
sigma = 0.02
kappa = 0.0000007 # risk aversion
tau = 5.0 # half-life
alpha = 50.0 # initial alpha strength
T = 30 # Total time steps

tau_list = np.linspace(2, 10, 20)
pnl_list = []
sharpe_list = []
for tau in tau_list:
    (deltas, positions, pnl_mean, pnl_std) = run(gamma, eta, beta, P, V, Theta, sigma, kappa, tau, alpha, T)
    pnl_list.append(pnl_mean)
    sharpe_list.append(np.sqrt(252)*pnl_mean/pnl_std)
    
plt.rcParams["figure.figsize"] = [15,7]
f, axarr = plt.subplots(1, 2)
axarr[0].plot(tau_list, pnl_list)
axarr[0].set_title('PnL')
axarr[0].set_xlabel('Tau')
axarr[1].plot(tau_list, sharpe_list)
axarr[1].set_title('Sharpe')
axarr[1].set_xlabel('Tau')
axarr[0].grid(True)
axarr[1].grid(True)
plt.show()
```
