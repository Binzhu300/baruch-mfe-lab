```python
import math
import numpy as np

%matplotlib inline
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings("ignore")

import pickle
```

```python
with open ('euput_rBergomi_results_H', 'rb') as fp:
    euput_rBergomi_results_H = pickle.load(fp)
with open ('amput_rBergomi_results_H', 'rb') as fp:
    amput_rBergomi_results_H = pickle.load(fp)
with open ('euput_localvol_results_H', 'rb') as fp:
    euput_localvol_results_H = pickle.load(fp)
with open ('amput_localvol_results_H', 'rb') as fp:
    amput_localvol_results_H = pickle.load(fp)
with open ('Klist_rBergomi_H', 'rb') as fp:
    Klist_rBergomi_H = pickle.load(fp)
with open ('Klist_localvol_H', 'rb') as fp:
    Klist_localvol_H = pickle.load(fp)
```

```python
euput_rBergomi_results_H = np.array(euput_rBergomi_results_H)
amput_rBergomi_results_H = np.array(amput_rBergomi_results_H)
euput_localvol_results_H = np.array(euput_localvol_results_H)
amput_localvol_results_H = np.array(amput_localvol_results_H)
```

```python
eep_rBergomi_results_H = amput_rBergomi_results_H - euput_rBergomi_results_H
eep_localvol_results_H = amput_localvol_results_H - euput_localvol_results_H
eep_differential_H = eep_rBergomi_results_H - eep_localvol_results_H
```
