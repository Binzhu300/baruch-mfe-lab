```python
import math
import numpy as np

%matplotlib inline
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings("ignore")

import pickle
```

```python
with open ('euput_rBergomi_results_H_fixed_eta', 'rb') as fp:
    euput_rBergomi_results_H_fixed_eta = pickle.load(fp)
with open ('amput_rBergomi_results_H_fixed_eta', 'rb') as fp:
    amput_rBergomi_results_H_fixed_eta = pickle.load(fp)
with open ('euput_localvol_results_H_fixed_eta', 'rb') as fp:
    euput_localvol_results_H_fixed_eta = pickle.load(fp)
with open ('amput_localvol_results_H_fixed_eta', 'rb') as fp:
    amput_localvol_results_H_fixed_eta = pickle.load(fp)
with open ('Klist_rBergomi_H_fixed_eta', 'rb') as fp:
    Klist_rBergomi_H_fixed_eta = pickle.load(fp)
with open ('Klist_localvol_H_fixed_eta', 'rb') as fp:
    Klist_localvol_H_fixed_eta = pickle.load(fp)
```

```python
euput_rBergomi_results_H_fixed_eta = np.array(euput_rBergomi_results_H_fixed_eta)
amput_rBergomi_results_H_fixed_eta = np.array(amput_rBergomi_results_H_fixed_eta)
euput_localvol_results_H_fixed_eta = np.array(euput_localvol_results_H_fixed_eta)
amput_localvol_results_H_fixed_eta = np.array(amput_localvol_results_H_fixed_eta)
```

```python
eep_rBergomi_results_H_fixed_eta = amput_rBergomi_results_H_fixed_eta - euput_rBergomi_results_H_fixed_eta
eep_localvol_results_H_fixed_eta = amput_localvol_results_H_fixed_eta - euput_localvol_results_H_fixed_eta
eep_differential_H_fixed_eta = eep_rBergomi_results_H_fixed_eta - eep_localvol_results_H_fixed_eta
```

```python
HList = np.linspace(0.0, 0.5, num=51)[3:]
len(HList)
```

```python
H_index = 2
print("H:", HList[H_index])
plt.plot(np.log(Klist_localvol_H_fixed_eta), eep_differential_H_fixed_eta[H_index])
```
