## Glosten and Milgrom

### 1. (8 points) 

(Adapted from Hasbrouck 5.2 but please note that the setup in (3) in the slides is different from Hasbrouck's in Figure 5.1 of his book.) Consider a variant of the Glosten and Milgrom model where there is informed trading only in the low state $V=\underline V$.

(a) Show that the learned probability after one step that the true price is $\bar V$ is given by

$$
\delta_1(Buy)=\Pr(\bar V|Buy)=\frac{\delta}{1-\mu\,(1-\delta)};\quad \delta_1(Sell)=\Pr(\bar V|Sell)=\frac{\delta}{1+\mu\,(1-\delta)}
$$

and that

$$
A=\frac{\underline V\,{(1-\delta)\,(1-\mu)}+\bar V\,{\delta}}{{1-\mu\,(1-\delta)}};\quad B=\frac{\underline V\,{(1-\delta)\,(1+\mu)}+\bar V\,{\delta}}{1+\mu\,(1-\delta)}
$$

(b) Deduce an expression for the spread $s=A-B$.

(c) Explain the sensitivities of the spread to the parameters of the model to a non-expert.


#### Solution (a):

$$
\delta_1(Buy) = \Pr(\bar V|Buy) = \frac{ \Pr(\bar V \cap Buy)}{ \Pr(Buy)} = \frac{\delta\times \frac{1}{2}}{\delta\times \frac{1}{2} + (1-\delta)\times\mu\times 0 + (1-\delta)\times(1-\mu)\times\frac{1}{2}} = \frac{\delta}{1-\mu(1-\delta)}.
$$

$$
\delta_1(Sell) = \Pr(\bar V|Sell) = \frac{ \Pr(\bar V \cap Sell)}{ \Pr(Sell)} = \frac{\delta\times \frac{1}{2}}{\delta\times \frac{1}{2} + (1-\delta)\times\mu\times 1 + (1-\delta)\times(1-\mu)\times\frac{1}{2}} = \frac{\delta}{1+\mu(1-\delta)}.
$$

$$
A = \mathbb{E}[V|Buy] = \Pr(\bar V|Buy)\bar V + \Pr(\underline V|Buy)\underline V = \frac{\delta \bar V + (1-\mu)(1-\delta)\underline V}{1-\mu(1-\delta)}.
$$

$$
B = \mathbb{E}[V|Sell] = \Pr(\bar V|Sell)\bar V + \Pr(\underline V|Sell)\underline V = \frac{\delta \bar V + (1+\mu)(1-\delta)\underline V}{1+\mu(1-\delta)}.
$$

#### Solution (b):

$$
s = A - B = \frac{\delta \bar V + (1-\mu)(1-\delta)\underline V}{1-\mu(1-\delta)} - \frac{\delta \bar V + (1+\mu)(1-\delta)\underline V}{1+\mu(1-\delta)} = \frac{2\mu (1-\delta)\delta(\bar V -\underline V)}{1 - \mu^2(1-\delta)^2}.
$$

#### Solution (c):

- $(\bar V -\underline V)$ can be regarded as an empirical proxy for price volatility. The bid-ask spread increases as the volatility increases.
- The spread is a monotonically increasing function of $\mu$. In other words, the higher the proportion of informed trader, the larger the spread. This is a consequence of compensating the adverse selection from the dealer's point of view.
- Write the spread as
$$
s = \frac{2\mu (\bar V -\underline V)}{ (1-\mu^2)\frac{1}{\delta} + \frac{1}{1-\delta} + \mu^2 }
$$
- When $\delta \ll 1$, the term $(1-\mu^2)\frac{1}{\delta}$ in the denominator dominates. Thus, the spread $s$ is an increasing function of $\delta$.
- When $\delta$ is close to 1, the term $\frac{1}{1-\delta}$ in the denominator dominates. Thus, the spread $s$ is an decreasing function of $\delta$.
- The critical point that $s$ reaches a maximum as a function of $\delta$ occurs at
$$
\delta^2 = (1-\mu^2)(1-\delta)^2 \Rightarrow \delta = \frac{\sqrt{1-\mu^2} - (1-\mu^2)}{\mu^2}.
$$

#### Solution (b): 

Lee and Ready tick test:
- If a trade is on an uptick, it is a buy;
- If the trade is on a downtick, it is a sell;
- If the trade is on a zero-uptick, it is a buy;
- If the trade is on a zero-downtick, it is a sell

```r
priceSeries <- as.numeric(tqdata$PRICE)
seriesLength <- length(priceSeries)
LeeReadyTradeSigns <- numeric(seriesLength-1)
```

```r
for(i in 2:length(priceSeries))
{
    if(priceSeries[i]>priceSeries[i-1])
        LeeReadyTradeSigns[i-1]=1
    else if(priceSeries[i]<priceSeries[i-1])
        LeeReadyTradeSigns[i-1]=-1
    else
        LeeReadyTradeSigns[i-1]=LeeReadyTradeSigns[i-2]
}

head(LeeReadyTradeSigns)
```

```r
cor(tradeSigns[2:seriesLength], LeeReadyTradeSigns)
```

#### Solution (d):

#### Solution (e):

Denote the tradeSigns series by $a$, and the LeeReadyTradeSigns series by $b$. Since the elements of $a$ and $b$ are all $\pm 1$'s, the ratio of trades correctly signed by the Lee-Ready test is
$$
p = \frac{\sum_i (1 + a_i b_i )}{2n} = \frac{1}{2}\left[ 1+\frac{\sum_i a_i b_i}{n} \right],
$$
where $n$ is the length of the sign series. Assume approximately $\frac{1}{n}\sum_i a_i \approx 0$, $\frac{1}{n}\sum_i b_i \approx 0$ as $n\gg 1$, the correlation coefficient 
$$
r = \frac{\sum_i a_i b_i - \sum_i a_i \sum_j b_j }{\sqrt{\sum_i a_i^2 \sum_j b_j^2}}
\approx \frac{\sum_i a_i b_i  }{n}.
$$
Thus,
$$
p \approx \frac{1+r}{2}.
$$
This is indeed the case, substitute $r=0.6618$, we get
$$
p \approx \frac{1+0.6618}{2} \approx 0.8309.
$$
