### 1. (10 points) 
Consider the first step $t = 1$ of the 3-step order book case presented in the lecture slides. Suppose that the order book at (or just before) $t = 1$ has one share offered and one share bid ($n_B = n_A = 1$).

Give the probabilities as of a market buy, market sell, limit buy and limit sell at $t = 1$,
You may use my Mathematica notebook *Parlour.nb* as a guide (if you wish).

<b>Solution</b>: Use the parameters given in the lecture example: $T=3$, $V=5.5$, $B=5$, $A=6$, and $\beta_t \in U(0,2)$. At $t=1$, we need to find the probability $\mathbb{P}^{B/S}_{t+1}$ of a fill (MB or MS) in period $t+1=2$. Because $n_A = n_B =1$, there is already a buy limit order in the front of the queue, one needs two successive market sell orders to come in for a buy limit order to be filled. Similar argument applies for the sell limit order. Using the derivation exemplified in lecture notes, $\mathbb{P}^B_{2}=\mathbb{P}^S_{2}=\left(\frac{5}{22}\right)^2=\frac{25}{484}$.

- $U_{MS} = B-\beta_t V = 5-5.5\beta_1$
- $U_{LS} = (A-\beta_t V)\mathbb{P}^B_{2} = \frac{25}{484}(6-5.5\beta_1)$
- $U_{LB} = (\beta_t V-B)\mathbb{P}^S_{2} = \frac{25}{484}(5.5\beta_1 -5)$
- $U_{MB} = \beta_t V - A = 5.5\beta_1 - 6$

<u>With probability $\frac{1}{4}$ the trader is a seller at $t=2$ and $t=3$</u>, 
- the trade is an MS if
$$
\begin{cases}
5-5.5\beta_1 > \frac{25}{484}(6-5.5\beta_1)\\
5-5.5\beta_1 > 0
\end{cases} 
\Rightarrow  \beta_1 < \frac{4540}{5049} \Rightarrow \text{Probability} = \frac{1}{4}\times \mathbb{P}\left(\beta_1 < \frac{4540}{5049}\right)  = \frac{1135}{5049} 
$$
- the trade is an LS if
$$
\begin{cases}
5-5.5\beta_1 < \frac{25}{484}(6-5.5\beta_1)\\
6-5.5\beta_1 > 0
\end{cases} 
\Rightarrow  \frac{4540}{5049} < \beta_1 < \frac{12}{11}
\Rightarrow \text{Probability} = \frac{1}{4}\times \mathbb{P}\left(\frac{4540}{5049} < \beta_1 < \frac{12}{11}\right)  = \frac{242}{5049} = \frac{22}{459}
$$

<u>With probability $\frac{1}{4}$ the trader is a buyer at $t=2$ and $t=3$</u> 
- the trade is an MB if
$$
\begin{cases}
5.5\beta_1 - 6 > \frac{25}{484}(5.5\beta_1 -5)\\
5.5\beta_1 - 6 > 0
\end{cases} 
\Rightarrow  \beta_1 > \frac{5558}{5049}
\Rightarrow \text{Probability} = \frac{1}{4}\times \mathbb{P}\left(\beta_1 > \frac{5558}{5049}\right)  = \frac{1135}{5049}

$$

### 2. (4 points) 
Explain why the optimal market/limit order decision depends in general on quantity available on both sides of the order book in the Parlour model.

<b>Solution</b>: Consider the case of a seller, the market order/limit order decision depends on
- the length of the queue on the ask side of the order book because the limit order at the front of the queue has priority of being filled;
- the length of the queue on the bid side of the order book because the longer the queue on the bid side, the more likely it is that a market buy order will be submitted in subsequent periods and the more likely that a limit sell order will be filled.

The same argument applies for the case of a buyer.

#### (d) Solution:

$$
\xi = \bar \xi \,\exp\left\{\sigma\,Z-\frac 12\,\sigma^2\right\}.
$$

with $Z \sim N(0,1)$ and $\sigma = 1/2$, is a log-normal variable.

- $\mathbb{E}[\xi] = \bar \xi$.
- $\mathbb{E}[\xi^2] = \bar{\xi}^2 e^{\sigma^2}$.

Thus, the variance $\mathbb{V}[\xi] = (e^{\sigma^2}-1)\bar{\xi}^2$. The standard deviation is $\sqrt{e^{\sigma^2}-1}\bar{\xi}=\sqrt{e^{0.5^2}-1}\bar{\xi} \approx 0.533 \bar\xi$.


#### (e) Solution:

According to lecture notes:
$$
    L ^ \star + Q = F^{-1} \left( \frac{2h+f+r}{\lambda_u + h+r} \right)
$$
where $F(x)$ denotes the cumulative distribution function of $\xi =\bar \xi \,\exp\left\{\sigma\,Z-\frac 12\,\sigma^2\right\}$. Thus,
$$
    L ^ \star + Q = \bar \xi \,\exp\left\{\sigma\,N^{-1}\left( \frac{2h+f+r}{\lambda_u + h+r} \right)-\frac 12\,\sigma^2\right\}
$$
where $N(x)$ denotes the cumulative distribution function of standard normal.

```r
optimalOrderSize <- function(volume, exchange)
{
    lambdaU = 3.75;
    h = 0.5;
    sigma = 0.5;
    if ( exchange == "BZX" ) {
        r = 0.32;
        f = 0.3;
    }
    else if ( exchange == "BYX") {
        r = -0.03;
        f = -0.03;
    }
    temp = qnorm( (2*h+f+r)/(lambdaU+h+r) ); 
    LplusQ = volume*exp(sigma*temp-0.5*sigma*sigma);
    return(LplusQ)
}

print(paste('Optimal order size traded on BZX =',optimalOrderSize(10500000*0.01,"BZX")))
print(paste('Optimal order size traded on BYX =',optimalOrderSize(1300000*0.01,"BYX")))
```

### 5. (6 points)

Another of Mike and Farmer's observations is that the distribution of marketable as well as unmarketable limit orders as a function of distance $\Delta$ to the same-side best quote is well-described by a Student-$t$ distribution.  Assume $\mu=1/2$ as above.

(a) Derive an expression in terms of the Student-$t$ cumulative density function for the ratio of arrival rates of limit to market orders as a function of the spread $s$.  How does this ratio vary with $s$?

(b) Relative to the zero-intelligence model, does this empirically observed strategic behavior make the spread revert faster or more slowly?

#### (a) Solution:

The ratio of arrival rates of limit orders to market orders $=\frac{\mathbb{P}(P>\text{bid})}{\mathbb{P}(P<\text{bid})} = \frac{F(s)}{F(-s)}$. This function is monotonically increasing with respect to $s$, where $F$ is the cumulative distribution function of Student-$t$.

#### (b) Solution:

Relative to the zero-intelligence model, the spread reverts faster. According to Part (a), the ratio of arrival rates of limit orders to that of market orders increases as the spread increases. As a consequence, more limited order arrives to facilitate the reversion of spread.
