### Our objective

$$
\newcommand{\BS}{\text{BS}}
$$

- Our objective is to approximate the shape of the implied volatility surface for any given stochastic volatility model, the Heston model in particular.

- In Lecture 2, we showed how to compute local volatilities from implied volatilities:

    <a name="eq1:localFromImplied"></a>(1)
$$
\begin{eqnarray*}
v_\ell(k,T) = \frac{\frac{\partial w}{\partial T}}
{\left(1 -\frac 12\frac{k}{w}\,\frac{\partial w}{\partial k}\right)^2
 - \frac14\,\left( \frac{1}{4} + \frac{1}{w} \right)\,
 \left( \frac{\partial w}{\partial k} \right)^2  +
\frac{1}{2}\,\frac{\partial^2 w}{\partial k^2 }}
\end{eqnarray*}
$$

- Recall that given a model, we may approximate the conditional expectation of $v_t$ to get $v_\ell$ using
    
    $$v_\ell(k,T)=\mathbb{E}[v_T\left|x_T=k\right.].$$  


- An obvious approach would then be to compute local volatility in the stochastic volatility model, then invert [(1)](#eq1:localFromImplied) to get implied volatility.

    - <span>We showed in Lecture 2 how to invert [(1)](#eq1:localFromImplied) for small times to
expiration.</span>

    - <span>However, nobody has yet figured out how to do this in general!</span>

- <span>Instead, by exploiting the insight (originally due to Bruno
Dupire) that implied volatility is related to the breakeven level of a
delta-hedging strategy, we derive a general path-integral representation
of Black-Scholes implied variance.</span>

### Delta hedging under stochastic volatility

Assume the stock price $S_t$ satisfies the SDE

$$\frac{{dS_t }}{{S_t }} = \mu _t dt + \sigma _{t } \,dZ_t$$ 

where the
volatility $\sigma _{t }$ may be random.
Then path-by-path, for any suitably smooth function $f\left( {S_t ,t}\right)$,

\begin{eqnarray*}
&&f\left( {S_T ,T} \right) - f\left( {S_0 ,0} \right) \\&=& \int_0^T
{df}\nonumber \\ &=& \int_0^T {\left\{ {\frac{{\partial
f}}{{\partial S_t }}\,dS_t  + \frac{{\partial f}}{{\partial
t}}\,dt + \frac{{\sigma _{t} ^2 }}{2}\, S_t ^2\,\frac{{\partial ^2
f}}{{\partial S_t ^2 }}\,dt} \right\}} \end{eqnarray*}

In particular (with $f(\cdot)=C_{BS}(\cdot)$), the value of a call
option is given by:

$$
\begin{eqnarray*}
&& C\left( {S_0,K,T} \right)\\ &=& \mathbb{ E}\left[ {\left( {S_T  -
K} \right)^+  } \right] \nonumber\\ &=&\mathbb{ E}\left[ {C_{BS} \left( {S_T, K,\bar\sigma(T),0}
\right)} \right] \nonumber \\ &=&
C_{BS} \left( {S_0,K ,\bar\sigma(0),T} \right)\nonumber\\
&& +
\mathbb{ E}\left[ { {\int_0^T {\left\{ {\frac{{\partial
C_{BS} }}{{\partial S_t }}dS_t  + \frac{{\partial C_{BS}
}}{{\partial t}}dt + \frac{1}{2}\sigma _{t} ^2 S_t ^2
\frac{{\partial ^2 C_{BS} }}{{\partial S_t ^2
}}dt} \right\}} }   } \right]
\end{eqnarray*}
$$

where $\bar \sigma(t),\,t \in [0,T]$ is some bounded positive function of $t$.

### The breakeven computation

Now let 

$$
\bar \sigma(t)^2 = \frac{1}{T-t}\, \int_t^T\,v(s)\,ds
$$

for some deterministic (bounded and non-negative) forward variance function $v(\cdot)$.  

Then 
$C_{BS} \left(S_t,K,\bar \sigma(t),T-t \right)$  satisfies the
Black-Scholes equation (assuming zero interest rates and dividends):

$$
\frac{{\partial C_{BS} }}{{\partial t}} =  -
\frac{1}{2}\,v(t)\,S_t ^2 \,\frac{{\partial ^2
C_{BS} }}{{\partial S_t ^2 }}.
$$ 

Substituting for
the time derivative $\frac{{\partial C_{BS} }}{{\partial t}}$, and using that $\mathbb{E}[dS_t]=0$, we
obtain:

$$
\begin{eqnarray*}
&&C\left( {S_0,K,T} \right)\\  
&=& C_{\BS} \left( {S_0,K ,\bar\sigma(0),T} \right) \nonumber\\&&+ \mathbb{
E}\left[ { {\int_0^T {\frac{1}{2}\left\{ \sigma _{t} ^2  -
v(t) \right\}S_t ^2 \frac{{\partial ^2
C_{BS} }}{{\partial S_t ^2 }}dt} }  }
\right]. \end{eqnarray*}
$$

In particular, if we choose $\bar \sigma(0) = \sigma_{\BS}(k,T)$, we have 
$
C\left( {S_0,K,T} \right) = C_{\BS} \left( {S_0,K ,\bar\sigma(0),T} \right) 
$
and

$$
\mathbb{
E}\left[ { {\int_0^T {\frac{1}{2}\left\{ \sigma _{t} ^2  -
v(t) \right\}S_t ^2 \frac{{\partial ^2
C_{BS} }}{{\partial S_t ^2 }}dt} }  }
\right]=0.
$$

### Representation of implied volatility

We then have the following lemma (from [Guyon and Henry-Labordère]<sup id="cite_ref-GHL" class="reference"><a href="#cite_note-GHL"><span>[</span>6<span>]</span></a></sup>  ):
<blockquote><div style="background-color:#add8e6; color:#FFFFFF; font-style: normal;  " ><h4>
Lemma 1 (Guyon and Henry-Labordère)</h4>
</div>
<div style="background-color:#E8E8E8; color:#000000; font-style: normal; ">
<br>
<a name="eq:lemma1"></a>(2)
$$\sigma_{BS}^2(K,T)=\frac{1}{T}\,\int_0^T\,v(t)\,dt$$ if and only if
$$\int_0^T 
{\mathbb{E} }\left[ \frac12\left\{ \sigma _t^2  -
v(t) \right\}S_t ^2\, \frac{{\partial}^2
C^v_{BS}}{{\partial}S_t ^2}\right]\,dt =0.$$ 


where $C^v_{BS}$ represents
the Black-Scholes formula computed with $v(t)$ as the forward variance
function. 


<br>
<br>

</div>

</blockquote>

<br>



### $\sigma_{BS}$ as the breakeven implied volatility

- In words, 
$$\int_0^T 
{\mathbb{E} }\left[ \frac12\left\{ \sigma _t^2  -
v(t) \right\}S_t ^2\, \frac{{\partial}^2
C^v_{BS}}{{\partial}S_t ^2}\right]\,dt $$ 
gives the
expected realized profit on the purchase of a European call or put option at the market price, delta-hedged using the deterministic
forward variance function $v(t)$ when the actual realized volatility
is $\sigma_t$. 

### Black-Scholes gamma

For fixed $K$ and $T$, define the Black-Scholes gamma

$$\Gamma_{\BS} (S_t,t): = \frac{{\partial ^2
}}{{\partial S_t ^2 }}\,C_{\BS}(S_t,K,\bar \sigma(t),T-t).$$ 

Then, in terms of $\Gamma_{\BS}$, the breakeven condition becomes

$$
\mathbb{
E}\left[ { {\int_0^T {\frac{1}{2}\left\{ \sigma _{t} ^2  -
v(t) \right\}S_t ^2 \,\Gamma_{\BS} (S_t,t )\,dt} }  }
\right]=0.
$$

### Black-Scholes forward implied variance

Now choose $v(t)$ to be the *Black-Scholes forward implied variance* function

<a name="eq:varformula"></a>(3)
$$v(t) = \frac{{\mathbb{ E}\left[ {\sigma _{t} ^2\, S_t ^2
\,\Gamma _{BS} (S_t,t )} \right]}}{{\mathbb{ E}\left[ {S_t ^2\, \Gamma _{BS}
(S_t,t )} \right]}}.
$$

Then

$${\mathbb{ E}\left[ {S_t ^2\, \Gamma _{BS} (S_t,t
)} \right]}\,v\left( t
\right) = {{\mathbb{ E}\left[ {\sigma _{t} ^2\, S_t ^2 \,\Gamma
_{BS} (S_t,t)}
\right]}}$$ 

and the breakeven condition [(4)](#eq:breakeven) is satisfied.


### $\sigma_{BS}^2$ as a path integral

It then follows from Lemma 1 [(2)](#eq:lemma1) that

<blockquote><div style="background-color:#add8e6; color:#FFFFFF; font-style: normal;  " ><h4>
Path integral representation of Black-Scholes implied volatility</h4>
</div>
<div style="background-color:#E8E8E8; color:#000000; font-style: normal; ">
<br>
<a name="eq:explicitimpvol"></a>(4)
$$\sigma_{BS}(K,T)^2 =\frac {1}{T} \int_0^T
{\frac{{\mathbb{ E}\left[ {\sigma _{t} ^2 S_t ^2 \Gamma _{BS} (S_t,t
)} \right]}}{{\mathbb{ E}\left[
{S_t ^2 \Gamma _{BS} (S_t ,t)}
\right]}}dt}$$

<br>
<br>

</div>

</blockquote>

- <span> Equation [(4)](#eq6:explicitimpvol) expresses implied variance as the
time-integral of expected instantaneous variance $\sigma_t^2$ under some
probability measure.</span>

- <span> [(4)](#eq6:explicitimpvol) is <span>*exact*</span> but implicit.</span>

### Interpretation


- To compute the Black-Scholes implied volatility of an option, we
need to average the possible realized volatilities over all possible
scenarios.



- Each such scenario is weighted by the gamma of the option.

    - The profitability of the delta-hedger in any time interval is
directly proportional to the gamma and the difference between “expected
instantaneous variance” (or local variance) and realized instantaneous
variance.

- Only paths that start at $S_0$ and end at the strike price $K$
need be included in the average.

    - At inception of the delta-hedge, the stock price can only be
$S_0$.

    - At expiration, gamma is precisely zero if $S_T \neq K$. Gamma
explodes at $S_T=K$.

### An elegant recasting of [(4)](#eq:explicitimpvol)

Following Roger Lee, we may rewrite [(4)](#eq:explicitimpvol) more elegantly
as

<a name="eq:gdefn"></a>(5)
$$\sigma_{BS}(K,T)^2=\bar\sigma(0)^2=\frac {1}{T}
\int_0^T\,\mathbb{E}^{G_t}[\sigma_t^2]\,dt$$

thus interpreting the definition [(5)](#eq:varformula) of $v(t)$ as the
expectation of $\sigma_t^2$ with respect to the probability measure
$\mathbb{G}_t$ defined, relative to the pricing measure $\mathbb{Q}$, by
the Radon-Nikodym derivative


$$\frac{d\mathbb{G}_t}{d\mathbb{Q}}:=\frac{{ { S_t ^2 \,\Gamma _{BS}
(S_t,\bar \sigma(t) )}}}{{\mathbb{ E}\left[ {S_t ^2\, \Gamma _{BS}
(S_t,\bar \sigma(t) )} \right]}}.$$

<font color="red">

#### Remark

Equations [(3)](#eq:varformula) and [(4)](#eq:explicitimpvol) are implicit
because the gamma $\Gamma
_{BS} (S_t ,t)$ of the option depends on all the forward implied variances
$v\left( t \right)$.

### Special Case: Black-Scholes

Suppose $\sigma _{t} =
\sigma(t)  $, a function of $t$ only. Then

$$v\left( t \right) = \frac{{\mathbb{ E}\left[ {\sigma (t) ^2
S_t ^2 \Gamma _{BS} (S_t )}
\right]}}{{\mathbb{ E}\left[ {S_t ^2 \Gamma _{BS} (S_t )} \right]}} = \sigma (t) ^2.$$

The forward implied variance $v(t)$ and the forward variance
$\sigma(t)^2$ coincide.

As expected,in this case, $v(t)$ has no dependence on the strike $K$ or the
option expiration $T$.

### Understanding the integral representation [(4)](#eq6:explicitimpvol)

In order to get better intuition for the path integral representation [(4)](#eq:explicitimpvol), first
recall how to compute a risk-neutral expectation:

$$\mathbb{ E}^{\mathbb{Q}}\left[ {f\left( {S_t } \right)} \right] = \int
{dS_t \,p\left( {S_t,t ;S_0 } \right)} \,f\left( {S_t } \right).$$ 

We get
the risk-neutral pdf of the stock price at time $t$ by taking the second
derivative of the market price of European options with respect to
strike price.

$$p\left( {S_t,t ;S_0 } \right) = \left. {\frac{{\partial ^2 C\left(
{S_0 ,K,t} \right)}}{{\partial K^2 }}} \right|_{K = S_t }.$$

Then from equation [(5)](#eq:gdefn) we have

<a name="eq6:brownianbridgeS"></a>(6)
\begin{eqnarray*}
v\left( t \right) &=&\mathbb{ E}^{{G_t}}\left[
\sigma_t^2  \right]\nonumber\\
&=& \mathbb{ E}^{\mathbb{Q}}\left[\sigma_t^2 \frac{d\mathbb{G}_t}{d\mathbb{Q}} \right]\nonumber\\
&=& \int {dS_t \,q\left( {S_t ;S_0 ,K,T} \right)} \,\mathbb{
E}^{\mathbb{Q}}\left[\left.\sigma _{t} ^2\right|S_t\right]\nonumber\\ &=& \int {dS_t
\,q\left({S_t ;S_0 ,K,T} \right)}
\,v_\ell(S_t,t)\end{eqnarray*}

where we further define

$$q\left( {S_t,t ;S_0 ,K,T} \right): = \frac{{\,p\left( {S_t,t;S_0 }
\right)\,S_t ^2\, \Gamma _{BS} (S_t ,t)}}{{\mathbb{ E}\left[ {S_t ^2
\,\Gamma _{BS} (S_t ,t)} \right]}}
$$

and
$v_\ell(S_t,t)=\mathbb{ E}^{\mathbb{Q}}\left[\left.\sigma _{t} ^2  \right|S_t\right]$ is
the local variance.

### Brownian Bridge-like density

- <span>$q\left( {S_t,t;S_0 ,K,T} \right)$ looks like a Brownian Bridge
density for the stock price:</span>

    - <span> $p\left(
{S_t,t;S_0 } \right)$ has a delta-function peak at $S_0$ at time $0$
and</span>

    - <span>$\Gamma _{BS} (S_t,t)$ has a delta-function peak at $K$ at
expiration $T$.</span>

For convenience in what follows, we now rewrite equation
[(6)](#eq6:brownianbridgeS) in terms of $x_t := \log {\left( S_t /
S_0 \right )}$:

<a name="eq7:brownianbridgex"></a>(7)
$$v\left(
t \right) = \int {dx_t \,q\left( {x_t ,t;x_T,T} \right)} \,v_\ell(x_t
,t) $$

### The quasi-Brownian Bridge density

```r
p <- function(St,t,S0,sigma){
    x <- log(St/S0)
    sig <- sigma*sqrt(t)
    return(dnorm(x/sig-sig/2)/sig/St)
   }
    
gammaBS <- function(S0,K,T,sigma){
    x <- log(S0/K)
    sig <- sigma*sqrt(T)
    d1 <- x/sig+sig/2
    return(dnorm(d1)/sig/S0)
   }

q <- function(St,t,S0,K,T,sigma){p(St,t,S0,sigma)*gammaBS(St,K,T-t,sigma)*St^2}

q1 <- function(x,t){q(exp(x),t,S0=1,K=1.3,T=1,sigma=.2)}
q1Vec <- function(k,t){sapply(t,function(t){q1(k,t)})} # Vectorized version for plotting
```

### The most-likely-path approximation

- <span>Figure 1 shows what $q\left( {x_t ,t;x_T,T} \right)$ looks
like in the case of a 1 year European option struck at 1.3 with a flat
20% volatility. </span>

    - <span>We see that $q\left( {x_t ,t;x_T,T} \right)$ peaks on a line
(which we will denote by $\tilde x_t$ and call the
<span>*most-likely-path*</span>) joining the stock price today with the
strike price at expiration.</span>

    - <span> Moreover, the density looks roughly symmetric around the peak.
This suggests an expansion around the peak $\tilde x_t$ (at which the
derivative of $q\left( {x_t ,t;x_t,T} \right)$ with respect to $x_t$ is
zero). </span>

This motivates us to write:

<a name="eq:approx"></a>(8)
$$q\left( {x_t ,t;x_T,T} \right) \approx q( \tilde x_t,t;x_T,T)
 + \frac{1}{2}\left( x_t  - \tilde
x_t \right)^2 \left. {\frac{\partial ^2 q}{\partial x_t ^2 }}
\right|_{x_t  = \tilde x_t} $$

In practice, the local variance $v_\ell(x_t,t)$ is typically not so far
from linear in $x_t$ in the region where $q\left( {x_t
,t;x_T,T} \right)$ is significant so we may further write

<a name="eq:sigapprox"></a>(9)
$$v_\ell(x_t,t) \approx v_\ell(\tilde x_t,t)
 + \left( x_t  - \tilde
x_t \right) \left. {\frac{\partial v_\ell }{\partial x_t }}
\right|_{x_t  = \tilde x_t} $$

Substituting [(8)](#eq:approx) and [(9)](#eq:sigapprox) into the integrand in
equation [(7)](#eq7:brownianbridgex) gives

$$v\left( t \right) \approx v_\ell(\tilde x_t,t)$$ 

and we may rewrite
equation [(4)](#eq:explicitimpvol) as

<a name="eq:impvolfromlocvol"></a>(10)
$$\sigma_{BS}(K,T)^2 \approx \frac {1}{T} \int_0^T {v_\ell(\tilde
x_t)\, dt}$$

### The MLP approximation in words

In words, equation [(10)](#eq:impvolfromlocvol) says that

<blockquote>
<div style="background-color:#E8E8E8; color:#000000; font-style: normal; ">
<br>
The Black-Scholes implied variance of an option with strike $K$ is given
approximately by the integral from valuation date ($t=0$) to the
expiration date ($t=T$) of the local variances along the path
$\tilde x_t$ that maximizes the Brownian Bridge density $q\left(
{x_t ,t;x_T,T} \right)$.

<br>
<br>

</div>

</blockquote>



- <span> Note that in practice, it’s not trivial to compute the most
likely path $\tilde
x_t$.</span>

    -   <span>[Adil Reghai]<sup id="cite_ref-Reghai" class="reference"><a href="#cite_note-Reghai"><span>[</span>7<span>]</span></a></sup> describes an efficient fixed-point algorithm to do
    this.</span>

### A fixed point algorithm for finding the most likely path

For each log-strike $k:=\log(K/S_0)$, we approximate the most likely
path $\tilde x_t$ as $$\tilde x_t = \frac{w_t}{w_T}\,k$$ with
$$w_t=\int_0^t\,ds \,{v_\ell(\tilde x_s)}$$

Obviously, this definition is circular. We solve by starting with the
straight line 

$$
\tilde x_t = \frac{t}{T}\,k
$$ 

as our initial guess and
iterating until the path doesn’t change. This iteration is extremely
fast in practice.

### Problems with the MLP approximation

- The MLP approximation should work well if local variance is
roughly linear in the neighborhood of the peak of $q(\cdot)$.

    - Conversely, if there is substantial curvature, the approximation
does not work well.
        - For example, at-the-money if there is curvature and little or no
skew (as in FX).

    - The MLP approximation is too rough for longer dates when the
density $q(\cdot)$ is not so highly peaked.

   - It’s not clear that the peak of the density is the appropriate most-likely-path. For example, Reghai actually approximates the conditional expectation:

$$\tilde S_t := {\mathbb{E} }\left[S_t|S_T=K\right]$$ 

- <span>[Guyon and Henry-Labordère]<sup id="cite_ref-GuyonHenryLabordere" class="reference"><a href="#cite_note-GuyonHenryLabordere"><span>[</span>5<span>]</span></a></sup> show how to correct for curvature (it is
rather slow and complicated).</span>

### Variational MLP

Later, [Gatheral and Wang]<sup id="cite_ref-GatheralWang" class="reference"><a href="#cite_note-GatheralWang"><span>[</span>4<span>]</span></a></sup> came up with another version of MLP where the
most-likely-path is now uniquely defined.

Define $$f(x(t),t)=\left.\partial_t \log \sigma(x,t)\right|_{x =x(t)}$$
Then
$$\sigma_{\text{BS}}(k,T) \approx \frac{\frac{1}{T}\,\int_0^T\,\sigma(\tilde x(t),t)\,\exp\left\{\int_0^t\,f(\tilde x(s),s)\,ds\right\}\,dt}
 {\sqrt{\frac{1}{T}\,\int_0^T\exp\left\{2\,\int_0^t\,f(\tilde x(s),s)\,ds\right\}\,dt}}.
$$ 

The most-likely-path $\tilde x(t)$ given (implicitly again) by

$$\tilde x(t)=k\,\frac{\int_0^t\,du\,\sigma(\tilde x(u),u)\,\exp \left\{\int_0^u\,f(\tilde x(s),s)\,ds\right\}}{\int_0^T\,du\,\sigma(\tilde x(u),u)\,\exp \left\{\int_0^u\,f(\tilde x(s),s)\,ds\right\}}.
$$

### Linear approximation to local variance

By definition,
$$
v_\ell(x,T) = \ee{v_T|x_T=x}.
$$

We may estimate this conditional expectation by linear regression:

$$
v_T = \xi_T(T) \approx \xi_0(T) + \beta(T)\,(x_T-\ee{x_T}) + \epsilon.
$$

Then

<a name="eq:v_ell"></a>(11)
$$
v_\ell(x,T) =\ee{v_T|x_T=x} \approx \xi_0(T) +  \beta(T)\,\left(x+ \frac12 w(T) \right).
$$

### Computation of the regression coefficient $\beta(T)$

$$
\beta(T) = \frac{\ee{Y_T\,M_T}}{\var[Y_T]}
$$

where $Y$ and $M$ are the martingales

$$
Y_T = \int_0^T\,\sqrt{v_t}\,dZ_t;\quad M_T = \int_0^T\,d\xi_t(T).
$$

### Affine forward variance models

From Lecture 3, a forward variance model has an affine CGF if and only if it takes the form

$$
\beas
\frac{dS_t}{S_t} &=& \sqrt{v_t}\,dZ_t\\
d\xi_t(u) &=& \sqrt{v_t}\, \kappa(u-t)\,dW_t
\eeas
$$

for some appropriate kernel $\kappa$ and with $\ee{dZ_t dW_t}= \rho\,dt$.

We may then compute

$$
\beas
\var[Y_T] &=& \ee{\int_0^T\,v_t\,dt} = w(T)\\
\ee{Y_T\,M_T} &=& \rho\,\ee{\int_0^T\,v_t\,\kappa(T-t)\,dt} = \rho\,\int_0^T\,\xi_0(t)\,\kappa(T-t)\,dt.
\eeas
$$

Thus
$$
\beta(T) = \frac{\rho}{w(T)}\,\int_0^T\,\xi_0(t)\,\kappa(T-t)\,dt.
$$


### Linear approximation to local variance in an affine model

For an affine forward variance model, we then have from [(11)](#eq:v_ell) that

<a name="eq:v_ell2"></a>(12)
$$
v_\ell(x,T)  \approx \xi_0(T)  + \rho\,\left(\frac x{w(T)}+\frac12\right)\,\int_0^T\,\xi_0(t)\,\kappa(T-t)\,dt.
$$


### Approximate local variance in the classical Heston model

In the classical Heston model 
$$ \kappa(T-t) = \eta\,e^{-\lambda(T-t)}
$$
and 
$$
\xi_0(t) = (v_0 - \bar v)\,e^{-\lambda t} + \bar v.
$$

Subsituting into [(12)](#eq:v_ell2) gives

<a name="eq:lvHeston"></a>(13)
$$
v_\ell(x,T)  \approx \xi_0(T) + \rho\,\eta\,\left(\frac x{w(T)}+\frac12\right)\,\int_0^T\,\xi_0(t)\,e^{-\lambda(T-t)}\,dt.
$$


### Flat forward variance curve

If $v_0=\bar v$, we have $\xi_0(t) = \bar v$.  In that case we obtain

$$
v_\ell(x,T)  \approx \bar v  + \rho\,\eta\,\left(\frac x T +\frac{\bar v}2\right)\,\int_0^T\,e^{-\lambda(T-t)}\,dt = \bar v + \rho\,\eta\,\left(x+\frac12 \,\bar v \,T\right)\,\frac{1-e^{-\lambda T}}{\lambda\,T}.
$$

- There is obviously something wrong with this approximation: It can go negative!
   - So we must floor it at zero.


### Approximate local variance in the rough Heston model

In the rough Heston model 
$$ \kappa(T-t) = \frac{\eta}{\Gamma(\alpha)}\,(T-t)^{\alpha-1}
$$
with $\alpha = H + \frac12$
and where
$
\xi_0(t)
$ is the infinite-dimensional state vector.

Subsituting into [(12)](#eq:v_ell2) gives

<a name="eq:lvRoughHeston"></a>(14)
$$
v_\ell(x,T)  \approx \xi_0(T) + \frac{\rho\,\eta}{\Gamma(\alpha)}\,\left(\frac x{w(T)}+\frac12\right)\,\int_0^T\,\xi_0(t)\,(T-t)^{\alpha-1}\,dt.
$$


### Flat forward variance curve

If  $\xi_0(t) = \bar v$, we obtain

$$
v_\ell(x,T)  \approx \bar v + \frac{\rho\,\eta}{\Gamma(\alpha+1)}\,\left(x+\frac12\,\bar v\,T\right)\,\frac{1}{T^{\frac12-H}}.
$$

- Note the power-law term structure of local volatility skew with exponent $\frac 12 - H$!

### A special case of classical Heston: The Heston-Nandi model

From equation [(13)](#eq:lvHeston), if $\rho=-1$ and if $v=\bar v$, Heston local variance should be well approximated by 

<a name="eq:HestonNandiLV"></a>(15)
$$
v_\ell(x,T)  \approx \left[ \bar v -\eta\,\left(x+\frac12 \,\bar v \,T\right)\,\frac{1-e^{-\lambda T}}{\lambda\,T}\right]^+.
$$

- The whole expression must be
bounded below by zero - all stock prices above the critical stock
price at which the local variance reaches zero are unattainable.

### Implied volatility in affine models

- Having computed approximate local variance in AFV models, we now apply the MLP appproximation to get implied variance.

    - <span> We need to integrate local variance along the most
likely stock price path joining the initial stock price to the strike
price at expiration (the one which maximizes the Brownian Bridge
probability density).</span> 

Then, the Black-Scholes implied variance is
given by

<a name="eq:impliedvariance"></a>(16)
$$\sigma_{BS}(K,T)^2 \approx \frac {1}{T} \int_0^T v_\ell({\tilde
x}_t,t)\, dt
$$

where $\left \{ \tilde x_t \right \}$ is the most likely path (as
defined earlier).

Recall that the Brownian Bridge density $q\left( {x_t ,t;x_T,T}
\right)$ is roughly symmetric and peaked around $\tilde x_t$, so $
\ee{ x_t - \tilde x_t\left| x_T \right.}
\approx 0$. This identifies the most likely path as

$$
\tilde x_t \approx \ee{ x_t\left|x_T \right.}.
$$

### The approximate most likely path

We do linear regression again.

$$
Y_t=x_t +\frac12 w(t)\approx \gamma(T)\,Y_T+\epsilon=\gamma(T)\,\left(x_T+\frac12 w(T)\right) + \epsilon. 
$$

with
$$
\gamma(T)= \frac{\ee{Y_t\,Y_T}}{\var{Y_T}}.
$$

Proceeding as before,

$$
\beas
\var{Y_T} &=& w(T)\\
\ee{Y_t\,Y_T} &=& w(t).
\eeas
$$

Thus
$$
\tilde x_t \approx \ee{ x_t\left|x_T \right.} \approx x_T\,\frac{w(t)}{w(T)}.
$$

- This is the same as the "ansatz" on page 32 of [The Volatility Surface].

We substitute this expression
back into equations [(12)](#eq:v_ell2) and [(16)](#eq:impliedvariance) to
get (noting that $x_T=k$),

<a name="eq:AffineImpliedVariance"></a>(17)
$$
\begin{eqnarray*}
\sigma_{BS}(k,T)^2 &\approx& \frac {1}{T}\int _0 ^T v_\ell(\tilde
x_t,t) dt \nonumber
\\
&\approx& \frac 1 T w(T)  + \frac{\rho}T\,\int_0^T\,\left(\frac {\tilde x_t}{w(t)}+\frac12\right)\,dt\,\int_0^t\,\xi_0(s)\,\kappa(t-s)\,ds\\
&\approx& \frac 1 T w(T)  + \frac{\rho}T\,\left(\frac {k}{w(T)}+\frac12\right)\,\int_0^T\,dt\,\int_0^t\,\xi_0(s)\,\kappa(t-s)\,ds\\
&=& \frac 1 T w(T)  + \frac{\rho}T\,\left(\frac {k}{w(T)}+\frac12\right)\,\int_0^T\,ds\,\xi_0(s)\int_0^{T-s}\,\kappa(r)\,dr.
\end{eqnarray*}
$$

### Special case: Flat forward variance curve

In the special case where $\xi_0(t)=\bar v$, we obtain

$$
\sigma_{BS}(k,T)^2 \approx \frac 1 T w(T)  + \frac{\rho}T\,\left(\frac {k}{\bar v T}+\frac12\right)\,\bar v\,\int_0^T\,ds\,\int_0^{T-s}\,\kappa(r)\,dr.
$$

### Implied volatility in the classical Heston model

In the classical Heston case, $\kappa(r) = \eta\, e^{-\lambda r}$.  Substituting back into 
equation [(17)](#eq:AffineImpliedVariance) gives

$$
\sigma_{BS}(k,T)^2 \approx \frac 1 T w(T)  + \frac{\rho\,\eta}{\lambda\,T}\,\left(\frac {k}{w(T)}+\frac12\right)\,\int_0^T\,ds\,\xi_0(s)\left(1- e^{-\lambda (T-s)}\right).
$$

In the special case that $v_0=\bar v$ (flat forward variance curve), we obtain

<a name="eq:HestonImpliedVariance"></a>(18)
$$
\sigma_{BS}(k,T)^2 \approx \bar v  + \frac{\rho\,\eta}{\lambda T}\,\left(k+\frac{\bar v T}2\right)\,\left\{1-\frac{1- e^{-\lambda T}}{\lambda T}\right\}.
$$


### Implied volatility in the rough Heston model

In the rough Heston case, 
$$\kappa(r) = \frac{\eta}{\Gamma(\alpha)}\, r^{\alpha-1}.$$
Substituting back into 
equation [(17)](#eq:AffineImpliedVariance) gives


$$
\sigma_{BS}(k,T)^2 \approx \frac 1 T w(T)  + \frac{\rho\,\eta}{\Gamma(\alpha+1)}\,\left(\frac {k}{w(T)}+\frac12\right)\,\frac 1 T \,\int_0^T\,ds\,\xi_0(s)\,(T-s)^\alpha.
$$

In the special case that $v_0=\bar v$ (flat forward variance curve), we obtain

<a name="eq:RoughHestonImpliedVariance"></a>(19)
$$
\sigma_{BS}(k,T)^2 \approx \bar v  + \frac{\rho\,\eta}{\Gamma(\alpha+2)}\,\left(k+\frac{\bar v T}2\right)\,T^{\alpha-1}.
$$


### The implied variance skew in affine models

- By inspection of [(17)](#eq:AffineImpliedVariance) the implied variance
skew in an affine model is

    - approximately linear in the correlation $\rho$ and the volatility of volatility $\eta$.

    - proportional to the double integral
    $$
    \int_0^T\,ds\,\xi_0(s)\int_0^{T-s}\,\kappa(r)\,dr.
    $$

- Thus the skew is
    - a sum of decaying exponentials under classical Heston.
    - a decaying power-law under rough Heston with the same exponent as the exponent of the kernel $\kappa$.


- When $\xi_0(t) = \bar v$, the variance skew is *independent* of the level of instantaneous
variance.

    - This remains approximately true even when the forward variance curve is not flat.

### Asymptotic analysis of the classical Heston variance skew

- The very short-dated skew is independent of $\lambda$ and $T$:

$$
\begin{eqnarray*}
\frac{\partial } {\partial x_t} \sigma_{BS}(K,T)^2 &=&\rho
\eta\, \frac{1}{{\lambda  T}}\left\{ {1 - \frac{{\left( {1 - e^{ -
\lambda T} } \right)}}{{\lambda T}}}
\right\}\\&\rightarrow&\,\frac{\rho \eta}{2}\textrm{ as } T
\rightarrow 0.
\end{eqnarray*}$$

- <span> The long-dated skew is inversely proportional to $T$:

$$
\begin{eqnarray*}
\frac{\partial } {\partial x_t} \sigma_{BS}(K,T)^2 &=&\rho
\eta\,
 \frac{1}{{\lambda T}}\left\{ {1 - \frac{{\left( {1 - e^{ -
\lambda T} } \right)}}{{\lambda T}}} \right\}\\
&\sim&\frac{\rho
\eta}{\lambda  T}\textrm{ as } T \rightarrow \infty.
\end{eqnarray*}
$$ </span>

- The skew is a monotonic decreasing function of time to expiry.

- Increasing $\eta$ causes the curvature of the implied
volatility skew (related to the kurtosis of the risk-neutral density) to
increase but we haven’t shown that here.

### Asymptotic analysis of the rough Heston variance skew

- The term structure of variance skew is trivially power-law.


- The skew is a monotonic decreasing function of time to expiry.
    - True for all affine models with monotonic decreasing kernels.

### On the accuracy of the affine implied variance approximation



- Equation [(17)](#eq:AffineImpliedVariance) gives us an approximation for implied variance in any affine single-factor model.
    - We plan to check this explicitly later in the lecture on Asymptotics of the volatility surface.

   
 - Equation [(17)](#eq:AffineImpliedVariance) agrees with the [Bergomi and Guyon]<sup id="cite_ref-BG" class="reference"><a href="#cite_note-BG">[1]</a></sup> (BG) implied volatility expansion to first order in $\eta$ and is thus exact to that order.
 
- However, Figure 6 demonstrates that the most-likely-path (MLP) approximation is far from perfect.
    - In general the MLP approximation is good for gaining intuition but not good enough for industrial applications.


 


   

### Dynamics of the volatility skew under local volatility

- Typically, in practice, the local variance skew
$\frac{\partial  v_{\ell}
}{\partial k}$ decays for large $T$.


- For example, the approximate skew formula
[(12)](#eq:v_ell2) for affine models gives 
<p>
$$
\frac{\p v_\ell(k,T)}{\p k}  \approx  \frac{\rho}{w(T)}\,\int_0^T\,\xi_0(t)\,\kappa(T-t)\,dt.
$$
<p>
which is of decreasing magnitude for large $T$ if the forward variance curve $\xi_0$ is reasonably flat.

- To get the forward volatility surface in a local volatility
model, we integrate over the local volatilities from the (forward)
valuation date to the expiration of the option along the most likely
path.

    - The forward volatility surface will be substantially flatter than
today’s because the forward local volatility skews are all
flatter.

- <span>Contrast this with a stochastic volatility model where implied
volatility skews are approximately time-homogeneous.</span>

    - Local volatility models imply that future BS implied volatility
surfaces will be flat relative to today’s surface.
<p>
<p>
    - In a stochastic volatility model, the shape of the surface depends only on the parameters, which are assume fixed.  Future BS implied volatility surfaces must look like today’s.

### The SVI parameterization

In the SVI (“Stochastic Volatility Inspired”) parameterization, for each
expiration, we write

<a name="eq:SVI"></a>(20)
$$\sigma^2_{BS}(k)=a+b\,\left\{\rho\,(k-m)+\sqrt{(k-m)^2+\sigma^2}\right\}
$$

where the coefficients $a$, $b$, $\rho$, $\sigma$ and $m$ depend on the
expiration.

- <span>SVI is just a <span>*parameterization*</span>, there are no underlying
dynamics associated with it. </span>

- <span>In [Gatheral and Jacquier]<sup id="cite_ref-GatheralJacquier" class="reference"><a href="#cite_note-GatheralJacquier"><span>[</span>3<span>]</span></a></sup>, we show that it is possible to guarantee no static
arbitrage when fitting SVI.</span>

    - <span>The functional form [(20)](#eq:SVI) is fitted to all expirations
simultaneously subject to the constraint that there should be no
calendar spread or butterfly arbitrage. </span>


- <span>Total implied variance may then be interpolated between slices to
give a smooth surface.</span>

    - <span> Here we use Stineman monotonic spline interpolation. </span>

### ATM skew in the classical Heston model

```r
varskewHeston <- function(paramVec,texp){
    
    rho.eta <- paramVec[1]
    lambda <- paramVec[2]

    lt <- lambda*texp
    varskew <- rho.eta/lt*(1-(1-exp(-lt))/lt)
    return(varskew)
    
    }
```

```r
varskewHeston(c(-.64*.39,1.15),texp)
```

Now fit this skew function to the empirical points, dropping the first one:

```r
skewObjective1 <- function(paramVec){
    
    varskewModel <- varskewHeston(paramVec,texp[-1])
    obj <-sum((varskewModel-varskew.empirical[-1])^2)
    return(obj*1000000)
    
    }
```

```r
skewObjective1(c(-0.7 * 0.39, 1.15))
```

```r
res1 <- optim(c(-0.7 * 0.39, 1.15),skewObjective1)
(fit1 <- res1$par )
```

Fit again dropping the first three points:

```r
skewObjective2 <- function(paramVec){
    
    varskewModel <- varskewHeston(paramVec,texp[-(1:3)])
    obj <-sum((varskewModel-varskew.empirical[-(1:3)])^2)
    return(obj*1000000)
    
    }

res2 <- optim(c(-.7*.39,1.15),skewObjective2)
(fit2 <- res2$par)
```

Now fit a power law:

```r
skewObjectivePL <- function(alpha){
    
    varskewModel <- varskew.empirical[5]*(texp[5]/texp[-1])^alpha
    obj <-sum((varskewModel-varskew.empirical[-1])^2)
    return(obj*1000000)
    
    }
```

```r
resPL <- optim(.4,skewObjectivePL)
(fitPL <- resPL$par)
```

### Build an objective function

```r
obj <- function(volData)function(paramVec){
    par.Heston <- list(
        lambda=paramVec[1],
        rho=paramVec[2],
        eta=paramVec[3],
        v=paramVec[4],
        vbar=paramVec[5])
    
    volData <- spxOptData
    volData$MidVol <- (volData$Bid+volData$Ask)/2
    volData <- volData[!is.na(volData$MidVol),]  #Eliminate zero bids
    volData$k <- log(volData$Strike/volData$Fwd)
    volData$VolSpread <- volData$Ask-volData$Bid
    
    iv <- Vectorize(impvolHeston(par.Heston))
    volData$ModelVol <- iv(volData$k,volData$Texp)
    volData$errSq <- ((volData$ModelVol - volData$MidVol)/volData$VolSpread)^2
    return(sum(volData$errSq))
    
 }    
```

```r
xi.curve.Heston <- function(params)function(t){
    
    lambda <- params$lambda
    v <- params$v
    vbar <- params$vbar
    
    res <- (v-vbar)*exp(-lambda*t)+vbar 
    return(res)

}
```

### Rough Heston implied vol formula

We use the rough Heston characteristic function rational approximation from Lecture 3.

```r
impvol.rHeston <- function(params.rHeston)function(k,t){
    
    phi.rHeston <- phiRoughHestonDhApprox(params.rHeston, xi.curve.20050915, 
                                          dh.approx= d.h.Pade33, n=100)
    return(impvol.phi(phi.rHeston)(k,t))
}
```

```r
params.rHeston <- list(H=0.14,eta=0.39,rho=-.65)
```

### Build an objective function

```r
obj.rHeston <- function(volData)function(paramVec){
    par.rHeston <- list(
        H=paramVec[1],
        eta=paramVec[2],
        rho=paramVec[3])
    
    volData <- spxOptData
    volData$MidVol <- (volData$Bid+volData$Ask)/2
    volData <- volData[!is.na(volData$MidVol),]  #Eliminate zero bids
    volData$k <- log(volData$Strike/volData$Fwd)
    volData$VolSpread <- volData$Ask-volData$Bid
    
    iv <- Vectorize(impvol.rHeston(par.rHeston))
    volData$ModelVol <- iv(volData$k,volData$Texp)
    volData$errSq <- ((volData$ModelVol - volData$MidVol)/volData$VolSpread)^2
    return(sum(volData$errSq))
    
 }    
```

### Summary of Lecture 6

- We showed how to approximate local volatility in the Heston model.
   - The approximation is accurate when $\rho = - 1$ (The Heston-Nandi model in particular) in which case we have a local volatility model that generates almost the same European option prices as the Heston model.
   

   
- The Black-Scholes implied variance of an option with strike $K$ is given
approximately by the integral from valuation date ($t=0$) to the
expiration date ($t=T$) of the local variances along the path
$\tilde x_t$ that maximizes a Brownian Bridge-like density.

   

-  We used this most-likely-path approximation to approximate implied volatility in affine stochastic volatility models.
    - We were then able to deduce the approximate term structures of ATM skew.
    - We swa that only rough volatility models generate the observed power-law behavior of ATM skew.
