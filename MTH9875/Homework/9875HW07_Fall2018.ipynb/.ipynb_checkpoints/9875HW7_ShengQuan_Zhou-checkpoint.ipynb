### 1. (8 points)

Consider the surface SVI (SSVI) parameterization of $w(k,\theta_t) = \sigma_{\BS}(k,t)^2\,t$:

<a name="eq:ssvi"></a>(1)
$$
w(k,\theta_t)
= \frac{\theta_t}{2}\left\{1+\rho\,\varphi(\theta_t)\, k + \sqrt{\left(\varphi(\theta_t)\,{k}+\rho\right)^2
+(1-\rho^2 )}\right\}.
$$

For economy of notation, write $w(k,\theta_t)=w$, $\varphi(\theta_t)=\varphi$ and so on.

(a)

Show that the ATM total implied variance $w(0,\theta) = \theta$, and that the ATM total variance skew $\p_k w|_{k=0}= \rho\,\theta\,\varphi$.

(b)

Find expressions for the minimum $w^\star$ of the SSVI total implied variance smile and the log-strike $k^\star$ at which the minimum is attained.


### Solution (a)

$$
w(0,\theta)=\frac{\theta}{2}\left\{1+0+\sqrt{\rho^2+1-\rho^2}\right\}=\theta.
$$
$$
\p_k w(k,\theta)|_{k=0}=\left.\frac{\theta}{2}\left\{\rho\varphi(\theta)+\frac{(\varphi(\theta)k+\rho)\varphi(\theta)}{\sqrt{(\varphi(\theta)k+\rho)^2+(1-\rho^2)}}\right\}\right|_{k=0} =\rho\theta\varphi(\theta).
$$

### Solution (b)

Let $\p_k w(k,\theta)=0$, we get
$$
\frac{1}{\phi(\theta)^2} + \frac{1-\rho^2}{(\phi(\theta)k+\rho)^2 \phi(\theta)^2} = \frac{1}{\rho^2\phi(\theta)^2}
\Rightarrow  \frac{1}{(\phi(\theta)k+\rho)^2 }  = \frac{1}{\rho^2} \Rightarrow k^\star =0 \text{ or } -\frac{2\rho}{\phi(\theta)}.
$$
Take $k^\star=-\frac{2\rho}{\phi(\theta)}$,
$$
w^\star = \theta(1-\rho^2).
$$

### Solution (a)

```r
paramsBCC <- list(lambda=1.15,rho=-0.64,eta=0.39,vbar=0.04,v=0.04)
closedFormImpVol<-function(k){sapply(k,function(k){ impvol.phi(phiHeston(paramsBCC))(k, 1.0) })}
```

```r
Ks<-seq(-1,1,0.01)
iVol_exact<-sapply(Ks,closedFormImpVol)

options(repr.plot.width=8, repr.plot.height=6)
plot(Ks,iVol_exact,type="l", lty=2, col="black", xlab="Strike K", ylab="Implied vol.")
```

### Solution (b)

### Solution (c)

With $\rho = -0.64$, let

\begin{eqnarray}
w^\star &=& \theta(1-\rho^2) \approx 0.0222391532881414,\\
k^\star &=& -\frac{2\rho}{\phi(\theta)} \approx 0.265324535680311,
\end{eqnarray}
we get
$$
\theta \approx 0.037668,\quad \phi \approx 4.824280.
$$

### Solution (d)

```r
SSVI_func<-function(k)
{
    rho = -0.64
    theta = 0.0376679425612151
    phi = 4.82428056160727
    return(sqrt(0.5*theta*(1+rho*phi*k+sqrt((phi*k+rho)*(phi*k+rho)+1-rho*rho))))
}

SSVI_ImpVol<-function(k) {sapply(k,function(k) {SSVI_func(k)})}
```

```r
Ks<-seq(-1,1,0.01)
iVol_exact<-sapply(Ks,closedFormImpVol)
iVol_SSVI<-sapply(Ks,SSVI_ImpVol)
plot(Ks,iVol_exact,type="l", lty=2, col="black", xlab="Strike K", ylab="Implied vol.")
lines(Ks,iVol_SSVI,col="red",lty=1)
```

### Answer (e)

No, SSVI does not fit the Heston smile in this parameter choice. The SVI parameterization is not arbitrary in the sense that the large-maturity limit of the Heston implied volatility smile is exactly SVI. We expect the SSVI formula to fit the Heston smile well in large-maturity regime.
