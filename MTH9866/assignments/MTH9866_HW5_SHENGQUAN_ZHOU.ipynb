### Question 1

In the stochastic spot/vol correlation model there are three correlations: the spot/vol correlation $\rho$ (which is stochastic); the spot/correlation correlation $\rho_{cs}$; and the vol/correlation correlation $\rho_{cv}$. 

In the formulation discussed in class, the latter two correlations are given as constant parameters. That 3x3 correlation matrix is not guaranteed to be positive definite, which is a problem for the model.

Show that, if the vol/correlation correlation is chosen as $\rho_{cv} = \rho\rho_{cs}$, the 3x3 correlation matrix is always positive definite.

#### Solution:

The correlation matrix is
$$
\begin{pmatrix}
1 & \rho & \rho_{cs}\\
\rho & 1 & \rho_{cv}\\
\rho_{cs} & \rho_{cv} &1
\end{pmatrix}.
$$

If $\rho_{cv} = \rho\rho_{cs}$, the leading principal minors are
\begin{eqnarray}
&& 1>0\\
&& 1-\rho^2 \ge 0\\
&& 1+ 2\rho\rho _ {cs}\rho _ {cv} - \rho^2 - \rho _ {cs}^2- \rho _ {cv}^2 = 1+ 2\rho^2\rho _ {cs}^2 - \rho^2 - \rho _ {cs}^2- \rho^2\rho _ {cs}^2 = 1+ \rho^2\rho _ {cs}^2 - \rho^2 - \rho _ {cs}^2 = (1-\rho^2)(1-\rho _ {cs}^2) \ge 0.
\end{eqnarray}
Thus, the 3x3 correlation matrix is always positive (semi-)definite.

### Question 2

For the Heston stochastic volatility model we showed that the characteristic function $f$ needed to satisfy the following partial differential equation in two variables, $x$ (the log of spot) and $v$ (the instantaneous volatility squared):
$$
\frac{\partial f}{\partial t} + \left(\mu -\frac{v}{2}\right)\frac{\partial f}{\partial x} + \frac{v}{2}\frac{\partial^2 f}{\partial x^2}  + \beta (\bar{v} -v)\frac{\partial f}{\partial v} + \frac{\alpha^2v}{2}\frac{\partial^2 f}{\partial v^2} + \rho\alpha v \frac{\partial^2 f}{\partial x\partial v} = 0
$$
and the initial condition for $f(t=T)$ is
$$
f = e^{i\theta x}.
$$
We guessed a form for $f$
$$
f = e^{i\theta x + A(t) + B(t)v}.
$$
Plug that form into the PDE above and show that it reduces the PDE to the two simultaneous ODEs
\begin{eqnarray}
\dot{A} &=& i\mu \theta + \beta \bar{v}B\\
\dot{B} &=& -\frac{1}{2}\left( i\theta + \theta^2 \right) -\beta B +\frac{1}{2}\alpha^2 B^2 + i\rho\alpha\theta B,
\end{eqnarray}
where $\dot{A} = \frac{dA}{d\tau}$ and $\dot{B} = \frac{dB}{d\tau}$, the new time variable $\tau = T-t$, and initial conditions are $A(0)=B(0)=0$.

#### Proof:

Just do the differentiations...

$$
\frac{\partial f}{\partial t} = -\frac{\partial f}{\partial \tau} = -f\left( \dot{A} + \dot{B}v \right),\\
\frac{\partial f}{\partial x} = if\theta, \quad \frac{\partial^2 f}{\partial x^2} = -f\theta^2,\\
\frac{\partial f}{\partial v} = fB, \quad \frac{\partial^2 f}{\partial v^2} = fB^2, \quad \frac{\partial^2 f}{\partial x\partial v} = i f\theta B.
$$
Then do the substitutions and factor away the common factor $f$...
$$
-\left( \dot{A} + \dot{B}v \right)+ i\left(\mu -\frac{v}{2}\right)\theta - \frac{v}{2} \theta^2 + \beta (\bar{v} -v)B + \frac{\alpha^2v}{2} B^2 + i\rho\alpha v \theta B = 0.
$$
Then group the terms
$$
\left( -\dot{A} + i\mu\theta + \beta \bar{v} B \right) + \left( -\dot{B}- \frac{i\theta}{2} -\frac{\theta^2}{2} - \beta B
+ \frac{\alpha^2 B^2}{2} + i\rho\alpha  \theta B\right)v = 0,
$$
for any $v$. Thus,
\begin{eqnarray}
\dot{A} &=& i\mu \theta + \beta \bar{v}B\\
\dot{B} &=& -\frac{1}{2}\left( i\theta + \theta^2 \right) -\beta B +\frac{1}{2}\alpha^2 B^2 + i\rho\alpha\theta B,
\end{eqnarray}
The initial conditions $A(0)=B(0)=0$ are obvious.

### Question 3

Consider a Merton jump diffusion model:

$$
dx(t) = \mu dt + \sigma dz + Jdq,
$$
where $x(t)$ is the log of spot, $\mu$ is the log-drift (a constant), $\sigma$ is the diffusive volatility of spot (a constant), and $q$ is a Poisson process with frequency $\lambda$. That is, in the infinitesimal time period $dt$, $dq$ is equal to 1 with probability $\lambda dt$ and equal to $0$ otherwise. $J$ is the jump that happens if the Poisson process fires; it is normally distributed with mean $J_a$ and variance $J_v$. Jumps are independent of each other, independent of the Poisson process, and independent of the Brownian motion $z$.

We’ll look at finding the characteristic function of $x(T)$ for some future time $T$, because with that we can calculate vanilla option prices with a single integration.

We can apply Ito’s Lemma to $f$, including the possibility of jumps, and get
$$
df = \frac{\partial f}{\partial t} dt + \frac{\partial f}{\partial x}dx + \frac{1}{2}\frac{\partial^2 f}{\partial x^2} dx^2 + \left[ f(x+J) - f(x) \right]dq.
$$
The last term is the one that comes from the jump $J$, which of course is random (drawn from its normal distribution).

As with the Heston model, let's try a form
$$
f = \mathbb{E}\left[ e^{i\theta x_T} \right] = e^{i\theta x + A(t)}.
$$
The question: substitute that form into the expression for $df$ and solve for $A(\tau =T-t)$. You will need to use the fact that $\mathbb{E}[df] = 0$ because the characteristic function is a martingale: but don't forget that the jump term has a non-zero contribution to that expectation.

You will need to use one of the most important identities in finance: for a normally distributed variable $u$, with mean $u_a$ and variance $u_v$,
$$
\mathbb{E}[e^u] = e^{u_a + \frac{u_v}{2}}.
$$
Your final answer should be
$$
A(\tau) = i\theta \mu \tau -\frac{\theta^2\sigma^2\tau}{2} + \lambda \tau \left( e^{i\theta J_a -\frac{1}{2}\theta^2 J_v} -1\right).
$$

#### Proof:

Do the differentiations:
$$
\frac{\partial f}{\partial t} = -f\dot{A}, \quad \frac{\partial f}{\partial x} = if\theta, \quad \frac{\partial^2 f}{\partial x^2} = -f\theta^2,
$$
Thus, using the Ito's lemma for jump processes (See Andrew Lesniewski MTH9876 Lecture 4)
$$
df(t,X_t) = \frac{\partial f(t,X_{t^-})}{\partial t}dt + \frac{\partial f(t,X_{t^-})}{\partial x} dX^c(t)
+ \frac{1}{2}\frac{\partial^2 f(t,X_{t^-})}{\partial x^2} dX^c(t)dX^c(t) + \Delta f(t,X_t)dJ(t),
$$
we get
\begin{eqnarray}
df &=& -f\dot{A}dt + if\theta dx - \frac{1}{2}f\theta^2 dx^2 + \left[ f(x+J) - f(x) \right]dq\\
&=& -f\dot{A}dt + if\theta(\mu dt + \sigma dz) - \frac{1}{2}f\theta^2\sigma^2 dt + \left[ f(x+J) - f(x) \right]dq\\
&=& \left( -f\dot{A} + if\theta\mu - \frac{1}{2}f\theta^2\sigma^2 \right)dt +   \left[  f(x+J) - f(x) \right]dq + if\theta \sigma dz
\end{eqnarray}
A convenient definition of a Martingale is $\mathbb{E}[df]=0$ (See Andrew Lesniewski MTH9876 Lecture 5, Page 22):
\begin{eqnarray}
0 &=& \mathbb{E}[df]\\
 &=& \mathbb{E}\left[ \left( -f\dot{A} + if\theta\mu - \frac{1}{2}f\theta^2\sigma^2 \right)dt +   \left[  f(x+J) - f(x) \right]dq + if\theta \sigma dz \right]\\
&=& \mathbb{E}\left[ \left( -f\dot{A} + if\theta\mu - \frac{1}{2}f\theta^2\sigma^2 \right)dt +   \left[  f(x+J) - f(x) \right]dq \right]\quad \text{Note: }\mathbb{E}[dz]=0\\
&=& \left( -f\dot{A} + if\theta\mu - \frac{1}{2}f\theta^2\sigma^2 \right)dt + \mathbb{E}\left[    f(x+J) - f(x) \right]\cdot \mathbb{E}\left[dq \right]\\
&=& \left( -f\dot{A} + if\theta\mu - \frac{1}{2}f\theta^2\sigma^2 \right)dt + \mathbb{E}\left[    f(x+J) - f(x) \right]\lambda dt, \quad \text{Note: } \mathbb{E}\left[dq \right]=\lambda dt\\
&=& \left( -f\dot{A} + if\theta\mu - \frac{1}{2}f\theta^2\sigma^2  + \lambda\mathbb{E}\left[   e^{i\theta (x+J) + A(t)} - e^{i\theta x + A(t)} \right] \right)dt, \quad \text{Note: } f = e^{i\theta x + A(t)}\\
&=& \left[ -f\dot{A} + if\theta\mu - \frac{1}{2}f\theta^2\sigma^2  + \lambda e^{i\theta x + A(t)} \left( \mathbb{E}\left[   e^{i\theta J}  \right]- 1 \right) \right]dt\\
&=& f\left[ -\dot{A} + i\theta\mu - \frac{1}{2}\theta^2\sigma^2  + \lambda  \left( e^{i\theta J_a -\frac{1}{2}\theta^2 J_v}- 1 \right) \right]dt.
\end{eqnarray}
Thus, we arrive at a first order ODE
$$
\dot{A} = i\theta\mu - \frac{1}{2}\theta^2\sigma^2  + \lambda  \left( e^{i\theta J_a -\frac{1}{2}\theta^2 J_v}- 1 \right),
$$
with an initial condition at $\tau = 0$ (or equivalently terminal condition at $t=T$):
$$
A(\tau = 0) = 0.
$$
Integrate once we get
$$
A(\tau) = i\theta \mu \tau -\frac{1}{2}\theta^2\sigma^2\tau + \lambda \tau \left( e^{i\theta J_a -\frac{1}{2}\theta^2 J_v} -1\right).
$$

### Question 4

Implement pricing for a jump diffusion model as in Question 3.

You should have two functions: one that prices a vanilla option using the characteristic function you derived in Question 3; and one that prices using conditional expectations, as defined below.

The first thing you need to do for either approach is figure out what drift $\mu$ to use to hit the forward to the expiration time $T$. For this, we need to calculate the value of a forward in the model; that is, the expected value of spot at time $T$.

For this we will use conditional expectations. That is, conditioned on exactly $N$ jumps happening by time $T$, $x$ is the sum of the normally-distributed Brownian motion plus the sum of the $N$ normally-distributed jumps; so, conditional on $N$ jumps, $x$ is normally distributed, and we can calculate the conditional expected value of spot:
$$
\mathbb{E}\left[ S(T)\mid N\text{ jumps} \right] = \mathbb{E}\left[ e^{x(T)} \right] = S(0)\mathbb{E}\left[e^{\mu T + \sigma z(T) + \sum_{i=1}^N J_i}\right].
$$
Complete this calculation. In calculating this expression, remember that the Brownian motion $z(T)$ and all the $N$ jumps $J_i$ are independent of each other.

For the next step, what we want is the unconditional expected value of $S(T)$, which we can identify with the forward (and then use to solve for the value of $\mu$ that makes the model match the forward). For this we can average over the number of jumps, weighting each by the probability of realizing that number of jumps. For a Poisson distribution, the probability of realizing $N$ jumps by time $T$ is given by
$$
\mathbb{P}(N) = \frac{(\lambda T)^N}{N!}e^{-\lambda T}.
$$
So to get the unconditional expected value of spot, equal to the forward, we can write
$$
F(T) = \mathbb{E}[S(T)] = \sum_{N=0}^{\infty} \mathbb{P}(N) \mathbb{E}\left[ S(T)\mid N\text{ jumps} \right].
$$
Calculate that quantity, which will determine the correct value of $\mu$ to use in your functions.

The second function should price the vanilla option using the same conditional expectation approach we just used to calculate the forward. That is, conditioned on $N$ jumps, the distribution of spot is lognormal, and we can calculate the conditional expected value and the conditional variance of log(spot). That means, conditioned on $N$ jumps you can calculate the conditional option price using the Black-Scholes formula. Then you can average over the number of jumps, as above, and get the unconditional option price.

You should implement the two functions and make sure that they give the same price for the same option. (Don’t forget to discount future cashflows!) For the numerical integration, use the scipy numerical integration package (eg scipy.integrate.quad).

Discuss the impact the four model parameters ($\sigma$, $J_a$, $J_v$, and $\lambda$) have on ATM vol, risk reversal, and butterfly.

#### Solution Step #1: Conditional expected value of forward
\begin{eqnarray}
\mathbb{E}\left[ S(T)\mid N\text{ jumps} \right] &=& S(0)\mathbb{E}\left[e^{\mu T + \sigma z(T) + \sum _ {i=1}^N J_i}\right]\\
&=& S(0)e^{\mu T } \mathbb{E}\left[e^{ \sigma z(T) }\right]\mathbb{E}\left[ e^{\sum _ {i=1}^N J_i}\right]\\
&=& S(0)e^{\mu T } \mathbb{E}\left[e^{ \sigma z(T) }\right]\Pi _ {i=1}^N \mathbb{E}\left[ e^{ J_i}\right]\\
&=& S(0)e^{\mu T }  e^{\frac{1}{2}\sigma^2 T} \Pi _ {i=1}^N e^{J_a + \frac{1}{2}J_v}\\
&=& S(0)e^{\mu T }  e^{\frac{1}{2}\sigma^2 T}  e^{N(J_a + \frac{1}{2}J_v)}\\
&=& S(0)e^{\mu T  + \frac{1}{2}\sigma^2 T + N(J_a + \frac{1}{2}J_v)}.
\end{eqnarray}
Then
\begin{eqnarray}
F(T) &=& \sum _ {N=0}^{\infty} \mathbb{P}(N) \mathbb{E}\left[ S(T)\mid N\text{ jumps} \right]\\
&=& S(0)\sum _ {N=0}^{\infty} \frac{(\lambda T)^N}{N!} e^{(\mu -\lambda) T  + \frac{1}{2}\sigma^2 T + N(J_a + \frac{1}{2}J_v)}\\
&=& S(0)e^{(\mu -\lambda) T  + \frac{1}{2}\sigma^2 T}\sum _ {N=0}^{\infty} \frac{(\lambda T e^{J_a + \frac{1}{2}J_v})^N}{N!} \\
&=& S(0)e^{(\mu -\lambda) T  + \frac{1}{2}\sigma^2 T} e^{\lambda T e^{J_a + \frac{1}{2}J_v}} \\
&=& S(0)e^{(\mu -\lambda) T  + \frac{1}{2}\sigma^2 T + \lambda T e^{J_a + \frac{1}{2}J_v}}.
\end{eqnarray}
Solve for $\mu$
$$
\mu = \frac{1}{T}\log\frac{F(T)}{S(0)} - \frac{1}{2}\sigma^2  - \lambda \left( e^{J_a + \frac{1}{2}J_v} - 1\right).
$$

#### Solution Step #2: Conditional variance of log-spot and volatility

\begin{eqnarray}
&& \mathbb{E}\left[ \log S(T) \mid N \text{ jumps} \right]
= \mathbb{E}\left[ \mu T + \sigma z(T) + \sum _ {i=1}^N J_i \mid N \text{ jumps} \right]
= \mu T + NJ_a = \mathcal{M} T \Rightarrow \mathcal{M} = \mu + \frac{NJ_a}{T}, \\
&&\mathbb{V}\left[ \log S(T) \mid N \text{ jumps} \right]
= \mathbb{V}\left[ \mu T + \sigma z(T) + \sum _ {i=1}^N J_i \mid N \text{ jumps} \right]
= \sigma^2 T + NJ_v = \Sigma^2 T \Rightarrow \Sigma = \sqrt{\sigma^2 + \frac{NJ_v}{T}}.
\end{eqnarray}

#### Solution Step #3: Link between Black-Scholes Option Pricing on Spot and Black Model on Forwards

The Black formula is similar to the Black–Scholes formula for valuing stock options except that the spot price of the underlying is replaced by a discounted futures price F.

The Black–Scholes formula for valuing stock options reads
\begin{eqnarray}
C(S_t, t) &=& +S_t e^{-q(T-t)}N(+d _ +) - Ke^{-r(T-t)}N(+d _ -),\\
P(S_t, t) &=& -S_t e^{-q(T-t)}N(-d _ +) + Ke^{-r(T-t)}N(-d _ -).
\end{eqnarray}
Recall that the forward price 
$$
F_t = S_t e^{(r-q)(T-t)},
$$
so that $F_0 = S_0 e^{(r-q)T}$ and $F_T = S_T$. In other words, $F_t$ has the same payoff as $S_t$ at $t=T$. The Black formula for valuing an option on a forward contract reads
\begin{eqnarray}
C(S_t, t) &=& +S_t e^{-q(T-t)}N(+d _ +) - Ke^{-r(T-t)}N(+d _ -) = e^{-r(T-t)}\left[+S_t e^{(r-q)(T-t)}N(+d _ +) - KN(+d _ -)\right]= e^{-r(T-t)}\left[+F_tN(+d _ +) - KN(+d _ -)\right],\\
P(S_t, t) &=& -S_t e^{-q(T-t)}N(-d _ +) + Ke^{-r(T-t)}N(-d _ -) = e^{-r(T-t)}\left[-S_t e^{(r-q)(T-t)}N(-d _ +) + KN(-d _ -)\right]= e^{-r(T-t)}\left[-F_tN(-d _ +) + KN(-d _ -)\right].
\end{eqnarray}
So the two formulas are really the same thing. See https://en.wikipedia.org/wiki/Black_model for some details.

The Black model is the pricing model for options on forwards – i.e. when we know the forward price of the asset rather than the spot. It’s just the Black-Scholes formula, expect that for the spot argument you pass in the forward, and both interest rates equal the denominated discount rate.

In this case, the forward is the conditional expected value of spot.

#### Solution Step #4: Conditional valuation of vanilla option

The conditional distribution (conditioned on exactly $N$ jumps by $T$) of the log-spot is a normal random variable with mean $\mu T  + NJ_a$ and variance $\sigma^2 T + NJ_v$, the conditional distribution of spot is lognormal. Use Black formula to price a vanilla option
$$
\text{Black}_N\left( F, K, T', \Sigma, r \right)
$$
with the following parameters:
- Forward price: $F=\mathbb{E}[S_T\mid N \text{ jumps}]$,
- Strike $K$,
- Expiration $T'$,
- Volatility: $\Sigma = \sqrt{\sigma^2 + \frac{NJ_v}{T}}$,
- Risk-neutral discount rate $r$.

The unconditional price

$$
V = \sum_{N=0}^{\infty} P(N)\times \text{Black}_N\left( F, K, T', \Sigma, r \right).
$$


#### Solution Step #5: Numerical Implementation and Comparison of Two Methods

Comments on the numerical results are given at the end.

```python
import cmath
import math
import scipy.integrate
import scipy.optimize
from scipy.stats import norm
```

```python
# Black-Scholes pricing formula and implied volatility
def bs_price(is_call,S,K,T,v,r,q):
    d1 = (math.log(S/K)+(r+v*v/2.)*T)/(v*math.sqrt(T))
    d2 = d1-v*math.sqrt(T)
    if is_call:
        price = S*math.exp(-q*T)*norm.cdf(d1)-K*math.exp(-r*T)*norm.cdf(d2)
    else:
        price = K*math.exp(-r*T)*norm.cdf(-d2)-S*math.exp(-q*T)*norm.cdf(-d1)
    return price

def bs_vega(is_call,S,K,T,v,r,q):
    d1 = (math.log(S/K)+(r+v*v/2.)*T)/(v*math.sqrt(T))
    return S * math.sqrt(T)*norm.pdf(d1)

def find_vol(target_value, is_call, S, K, T, r, q):
    MAX_ITERATIONS = 5000
    PRECISION = 1.0e-6

    sigma = 0.5
    for i in range(0, MAX_ITERATIONS):
        price = bs_price(is_call, S, K, T, sigma, r, q)
        vega = bs_vega(is_call, S, K, T, sigma, r, q)

        price = price
        diff = target_value - price

        #print(i, sigma, diff)
        if (abs(diff) < PRECISION):
            return sigma
        
        sigma = sigma + diff/vega # f(x) / f'(x)

    return sigma
```

```python
def diffusive_drift(S, F, T, vol, lam, Ja, Js):
    '''
    S:   current spot of the asset
    F:   current forward price of the asset for settlement at texp
    T:   time to settlement
    vol: diffusive spot volatility
    lam: jump frequency
    Ja:  mean jump size in log(spot)
    Js:  standard deviation of the jump size in log(spot)
    '''
    
    return math.log(F/S)/T - lam * (math.exp(Ja + 0.5 * Js * Js) - 1) - 0.5 * vol * vol
```

```python
class MertonJumpDiffusionOptionPricer:
    
    def __init__(self, spot, vol, rd, rf, lam, Ja, Js):
        self.spot = spot
        self.vol = vol
        self.rd = rd
        self.rf = rf
        self.lam = lam
        self.Ja = Ja
        self.Js = Js
        
    def price_by_characteristic_function(self, strike, expiry, is_call):
        
        fwd = self.spot * math.exp((self.rd - self.rf) * expiry)
        mu = diffusive_drift(self.spot, fwd, expiry, self.vol, self.lam, self.Ja, self.Js)

        i = 1j
        Jv = self.Js * self.Js
        x = math.log(self.spot)

        def characteristic_function(theta):
            A = i * theta * mu * expiry - 0.5 * theta * theta * self.vol * self.vol * expiry + self.lam * expiry * (cmath.exp(i * theta * Ja - 0.5 * theta * theta * Jv) - 1)
            return cmath.exp(i * theta * x + A)

        strike_ratio = math.log(strike / self.spot)

        def integrand(theta):
            f = characteristic_function(theta)
            val = f * cmath.exp(-i * theta * strike_ratio) / (theta * theta + i * theta)
            return val.real

        theta_scale = math.sqrt(2 / self.vol / self.vol / expiry)

        integral = 0
        theta_lo = 0
        theta_hi = theta_scale
        count    = 0
        while True:
            integral_piece = scipy.integrate.quad(integrand, theta_lo, theta_hi)[0]
            integral += integral_piece
            if abs(integral_piece) < 1e-10: 
                break

            theta_lo  = theta_hi
            theta_hi += theta_scale

            count += 1
            if count > 1000: 
                raise ValueError('Numerical integration not converging')

        price = fwd - 0.5 * strike - (strike / math.pi) * integral
        if not is_call:
            price -= fwd - strike

        price *= math.exp(-rd * expiry)
        return price
    
    def price_by_conditional_expectation(self, strike, expiry, is_call):

        fwd = self.spot*math.exp((self.rd - self.rf) * expiry)
        mu = diffusive_drift(self.spot, fwd, expiry, self.vol, self.lam, self.Ja, self.Js) + 0.5 * self.vol * self.vol

        price = 0
        n_jumps = 0
        fact = 1
        elt = math.exp(-self.lam * expiry)

        Jv = self.Js * self.Js
        while True:
            prob_n_jumps = elt * math.pow(self.lam * expiry, n_jumps) / fact

            fwd_cond = self.spot * math.exp(mu * expiry + n_jumps * (self.Ja + 0.5 * Jv))
            var_cond = self.vol * self.vol * expiry + n_jumps * Jv
            vol_cond = math.sqrt(var_cond / expiry)

            price_cond = bs_price(is_call, fwd_cond, strike, expiry, vol_cond, 0, 0)
            price_piece = price_cond * prob_n_jumps
            price += price_piece
            if abs(price_piece / strike) < 1e-14: 
                break

            fact *= n_jumps + 1
            n_jumps += 1

        price *= math.exp(-rd * expiry)
        return price
    
    def implied_vol_by_characteristic_function(self, strike, expiry):

        fwd = self.spot*math.exp((self.rd - self.rf) * expiry)
        is_call = strike >= fwd

        price = self.price_by_characteristic_function(strike, expiry, is_call)
        vol = find_vol(price, is_call, self.spot, strike, expiry, self.rd, self.rf)
        return vol
    
    def implied_vol_by_conditional_expectation(self, strike, expiry):

        fwd = self.spot * math.exp((self.rd - self.rf) * expiry)
        is_call = strike >= fwd

        price = self.price_by_conditional_expectation(strike, expiry, is_call)
        vol = find_vol(price, is_call, self.spot, strike, expiry, self.rd, self.rf)
        return vol
```

```python
spot = 1
pnts = 0.03
expiry = 0.5
vol = 0.08
rd = 0.05
lam = 3
Ja = -0.04
Js = 0.

fwd = spot + pnts
rf = rd - 1 / expiry * math.log(fwd / spot)
delta_limit = 0.10
```

```python
pricer = MertonJumpDiffusionOptionPricer(spot, vol, rd, rf, lam, Ja, Js)
```

```python
strike_hi = 0.5
strike_lo = 2.0
nstrikes = 100
dstrike  = (strike_hi - strike_lo) / (nstrikes - 1)
strikes = []
vols_characteristic_function = [] 
vols_conditional_expectation = []

for i in range(nstrikes):
    strike = strike_lo + i*dstrike
    strikes.append(strike)

    vol_characteristic_function = pricer.implied_vol_by_characteristic_function(strike, expiry)
    vols_characteristic_function.append(vol_characteristic_function * 100)

    vol_conditional_expectation = pricer.implied_vol_by_conditional_expectation(strike, expiry)
    vols_conditional_expectation.append(vol_conditional_expectation * 100)
```

```python
import matplotlib.pyplot as plt
line1, = plt.plot(strikes, vols_characteristic_function, label='Method #1: Characteristic Function')
line2, = plt.plot(strikes, vols_conditional_expectation, label='Method #2: Conditional Expectation')
plt.legend(handles=[line1, line2])
plt.xlabel('Strikes')
plt.ylabel('Implied Vol (%)')
plt.title('Calculation by Two Methods')
plt.grid(True)
plt.show()
```

#### Comments

Smile can be observed in the above chart.

ATM volatility is determined by the interaction of three parameters:
- The lower bound: $\sigma$
- Jump mean $J_a$
- Jump variance $J_v$.

The impact of the jumps is larger for larger tenors.

The risk reversal (skew) is determined primarily by the jump mean $J_a$. 
\begin{eqnarray}
J_a > 0 \Rightarrow \text{ positive risk reversal }\\
J_a < 0 \Rightarrow \text{ negative risk reversal }
\end{eqnarray}

The butterfly (smile) is determined primarily by the jump variance $J_v$: 
$$
J_v\nearrow \Rightarrow \text{ butterfly }\nearrow.
$$
